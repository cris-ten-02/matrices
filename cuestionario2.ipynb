```python
def calcular_saldo(n_meses):
    # Variables del problema
    capital_inicial = 2000000
    factor_crecimiento = 1.012
    
    # Aplicamos la fórmula del interés compuesto: S_n = S_0 * (1.012)^n
    saldo_final = capital_inicial * (factor_crecimiento ** n_meses)
    
    return saldo_final

# Ejemplo: Calcular el saldo después de 5 meses
meses = 5
saldo_5_meses = calcular_saldo(meses)

print(f"El saldo después de {meses} meses es: ${saldo_5_meses:,.2f}")

```

```python
# Datos conocidos
a_2 = 1066  # Rendimiento al segundo mes
a_3 = 2132  # Rendimiento al tercer mes
n_objetivo = 9  # Queremos encontrar el rendimiento del mes 9

# 1. Calcular la razón constante (d)
# En una progresión aritmética, d = a_n - a_{n-1}
razon_constante = a_3 - a_2
print(f"La razón constante (d) de crecimiento es: {razon_constante} TPS/mes")
# razón_constante = 1066


# 2. Calcular el primer término (a_1)
# Usando a_2: a_1 = a_2 - d
a_1 = a_2 - razon_constante
# a_1 = 1066 - 1066 = 0
print(f"El rendimiento del primer mes (a_1) es: {a_1} TPS")

def rendimiento_mes_n(n, a_inicial, d):
    """
    Calcula el rendimiento (a_n) usando la fórmula de la progresión aritmética:
    a_n = a_1 + (n - 1) * d
    """
    rendimiento = a_inicial + (n - 1) * d
    return rendimiento

# 3. Calcular el rendimiento al noveno mes (a_9)
a_9 = rendimiento_mes_n(n_objetivo, a_1, razon_constante)

print("-" * 40)
print(f"El rendimiento del servidor el mes {n_objetivo} es: {a_9} TPS")
print("-" * 40)
```

```python
# Variables conocidas del problema
a_n_objetivo = 34112  # Rendimiento objetivo (a_n)
a_1 = 0              # Rendimiento inicial (a_1)
razon_constante = 1066 # Razón constante (d)

# 1. Aplicar la fórmula despejada para n
# n = ( (a_n - a_1) / d ) + 1

# Calculamos el término (n - 1) primero
n_menos_1 = (a_n_objetivo - a_1) / razon_constante

# Calculamos n sumando 1
n_final = n_menos_1 + 1

# Convertimos el resultado a un entero (ya que el mes debe ser un número entero)
mes_resultado = int(n_final)

print("-" * 40)
print(f"Rendimiento Objetivo (TPS): {a_n_objetivo}")
print(f"Razón Constante (d): {razon_constante}")
print("-" * 40)
print(f"El rendimiento de {a_n_objetivo} TPS se alcanza en el mes: {mes_resultado}")
print("-" * 40)
```

```python
# Datos conocidos
a_4 = 1108000  # Arriendo del mes 4
a_5 = 1144000  # Arriendo del mes 5
arriendo_objetivo = 1828000 # Arriendo que queremos encontrar el mes (a_n)

# 1. Calcular la razón constante (d)
razon_constante = a_5 - a_4
# d = 1144000 - 1108000 = 36000

print(f"Razón Constante (Incremento Mensual): ${razon_constante:,.0f}")


# 2. Calcular el primer término (a_1)
# n=4, por lo tanto (n-1) = 3
a_1 = a_4 - (3 * razon_constante)
# a_1 = 1108000 - (3 * 36000) = 1000000

print(f"Arriendo del Primer Mes (a_1): ${a_1:,.0f}")

# 3. Calcular el mes (n) para el arriendo objetivo
# Usamos la fórmula despejada para n
# n = ((a_n_objetivo - a_1) / d) + 1

n_menos_1 = (arriendo_objetivo - a_1) / razon_constante
# n_menos_1 = (1828000 - 1000000) / 36000 = 828000 / 36000 = 23.0

mes_resultado = int(n_menos_1 + 1)
# mes_resultado = 23 + 1 = 24

print("-" * 40)
print(f"Mes en que el arriendo es ${arriendo_objetivo:,.0f}: Mes {mes_resultado}")
print("-" * 40)
```

```python
# T: Matriz de unidades producidas (Plantas x Bodegas)
# Fila 2 corresponde a la Planta 2
T = [
    [300, 272, 240], # Planta 1
    [260, 180, 324]  # Planta 2
]

# U: Matriz de costo de almacenamiento (Bodegas x Meses)
# Columna 2 (índice 1) corresponde al mes de Febrero
U = [
    [1200, 1250], # Bodega 1
    [1100, 1400], # Bodega 2
    [1350, 1200]  # Bodega 3
]

# Identificadores
planta_idx = 1  # Planta 2 está en el índice 1 (segunda fila de T)
mes_idx = 1     # Febrero está en el índice 1 (segunda columna de U)

# Unidades producidas por la Planta 2 para cada bodega (fila de T)
unidades_planta_2 = T[planta_idx]
# [260, 180, 324]

# Costo unitario de almacenamiento en Febrero para cada bodega (columna de U)
costos_febrero = [U[i][mes_idx] for i in range(len(U))]
# [1250, 1400, 1200]

costo_total = 0

# Iteramos a través de las bodegas (índices 0, 1, 2)
for i in range(len(unidades_planta_2)):
    unidades = unidades_planta_2[i]
    costo_unitario = costos_febrero[i]
    
    # Acumulamos el costo
    costo_total += unidades * costo_unitario

print("-" * 50)
print("Desglose del Cálculo:")
print(f"Bodega 1: {unidades_planta_2[0]} * {costos_febrero[0]} = {unidades_planta_2[0] * costos_febrero[0]:,.0f}")
print(f"Bodega 2: {unidades_planta_2[1]} * {costos_febrero[1]} = {unidades_planta_2[1] * costos_febrero[1]:,.0f}")
print(f"Bodega 3: {unidades_planta_2[2]} * {costos_febrero[2]} = {unidades_planta_2[2] * costos_febrero[2]:,.0f}")
print("-" * 50)
print(f"Costo Total de almacenamiento (Planta 2, Febrero): {costo_total:,.0f}")
print("-" * 50)
```
