```python

# REGION        1      2   3       4 
A = np.array([[2.6 , 4.8 , 1.8 , 0.9], # producto 1 
             [3.2 , 4.4 , 2.5 , 2.8],  # producto 2
             [2.4 , 3.6 , 3.8 , 2.5]])  # producto 3

B = np.array([[3.6,2.5,3.0,2.5],
             [4.5,5.0,3.5,3.8],
             [2.9,3.0,4.6,4.0]])
S = A + B
D = B - A
print (f'S(3x4) =  \n {S}')
print (f'el total de las ventas en los dos años del producto 2 en la region 4 es  s24 = {S[1,3]} mill dolares ')
print(f'D(3x4= \n {D}')
print (f'el AUMENTO  de las ventas en los dos años del producto 2 en la region 4 es  S24 = {D[1,3]} mill dolares ')

print (f'LA DISMINUCION   de las ventas en EL 2DO AÑO RESPECTO AL PRIMERO  en la region 2 es  D32 = {D[2,1]:.1f} mill dolares ')
```
