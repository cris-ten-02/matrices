# ¿Por qué no se puede multiplicar Q * H? (Explicación sencilla)

Para multiplicar dos matrices, el número de columnas de la primera debe ser igual al número de filas de la segunda.

Por ejemplo:

Supongamos que Q es de tamaño 2x3 (2 filas, 3 columnas) y H es de tamaño 2x2 (2 filas, 2 columnas):

Q = | 1  2  3 |
    | 4  5  6 |

H = | 7  8 |
    | 9 10 |

No se puede multiplicar Q * H porque Q tiene 3 columnas y H solo tiene 2 filas. ¡No coinciden!

Pero si H fuera de tamaño 3x2 (3 filas, 2 columnas):

H = | 7  8 |
    | 9 10 |
    |11 12 |

Ahora sí se puede multiplicar Q * H porque Q tiene 3 columnas y H tiene 3 filas.

```python
import numpy as np
# Ejemplo sencillo de multiplicación válida
Q = np.array([[1, 2, 3],
              [4, 5, 6]])  # 2x3
H = np.array([[7, 8],
              [9, 10],
              [11, 12]])  # 3x2
resultado = Q @ H
print("Resultado de Q * H:")
print(resultado)
```

```python
# Ejemplo de multiplicación inválida (esto da error)
Q = np.array([[1, 2, 3],
              [4, 5, 6]])  # 2x3
H = np.array([[7, 8],
              [9, 10]])    # 2x2
# Esto generaría un error:
# resultado = Q @ H
# print(resultado)
# Porque 3 columnas de Q ≠ 2 filas de H
```
